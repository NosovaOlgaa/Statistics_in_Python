Справочная информация:

surgery? (1 = Yes, it had surgery, 2 = It was treated without surgery)  
Age (1 = Adult horse, 2 = Young (< 6 months))  
rectal temperature (normal temp is 37.8)  
pulse (30 - 40 is normal for adults)  
respiratory rate (normal rate is 8 to 10)  
temperature of extremities (1 = Normal,2 = Warm,3 = Cool, 4 = Cold)  
pain (1 = alert, no pain, 2 = depressed, 3 = intermittent mild pain, 4 = intermittent severe pain, 5 = continuous severe pain)    
outcome (1 = lived, 2 = died, 3 = was euthanized)

```python
horse_data.isna().sum()

#Кол-во пропусков в представленном df
```

```python
horse_data[horse_data['surgery?'].isnull()]

#данные по наличию операции и исходу пропущены в одной строке.
#т.к. остальные строки по этим параметрам заполнены, принимаю решение заполнить пропуски значением моды.
```

```python
horse_data['surgery?'] = horse_data['surgery?'].fillna(horse_data['surgery?'].mode()[0])
horse_data['outcome'] = horse_data['outcome'].fillna(horse_data['outcome'].mode()[0])

#посмотрим, какими значениями заполнились пропуски
horse_data.loc[horse_data['rectal_temperature'] == 38].head(10)
```

```python
horse_data[horse_data['pain'].isnull()]
```

```python
#предположу, что исход лошади взаимосвязан с уровнем боли, которую она испытывает при болезни.
#исходя из этого предположения, заполняю столбец pain медианой

horse_data['pain'] = horse_data['pain'].fillna(horse_data.groupby('outcome')['pain'].transform('median'))

#посмотрим, какими значениями заполнились пропуски
horse_data.loc[horse_data['outcome'] == 2].head(10)
```

```python
horse_data[horse_data['temperature_of_extremities'].isnull()]
```

```python
#предположу, что температура конечностей лошади связана с уровнем боли.
#исходя из этого предположения, заполняю столбец temperature_of_extremities медианой

horse_data['temperature_of_extremities'] = horse_data['temperature_of_extremities'].fillna(horse_data.groupby('pain')['temperature_of_extremities'].transform('median'))

#посмотрим, какими значениями заполнились пропуски
horse_data.loc[horse_data['pain'] == 3].head(10)
```

```python
horse_data[horse_data['pulse'].isnull()]
```

```python
#думаю, пульс можно заполнить в сцепке с уровнем боли

horse_data['pulse'] = horse_data['pulse'].fillna(horse_data.groupby('pain')['pulse'].transform('median'))

#посмотрим, какими значениями заполнились пропуски
horse_data.loc[horse_data['pain'] == 3].head(10)
```

```python
horse_data[horse_data['respiratory_rate'].isnull()]
```

```python
#частота дыхания зависит от пульса

horse_data['respiratory_rate'] = horse_data['respiratory_rate'].fillna(horse_data.groupby('pulse')['respiratory_rate'].transform('median'))

#посмотрим, какими значениями заполнились пропуски
horse_data.loc[horse_data['pulse'] == 60].head()
```

```python
horse_data[horse_data['rectal_temperature'].isnull()]
```

```python
#в свою очередь, ректальная температура, скорее всего, находится во взаимосвязи с пульсом и частотой дыхания.
#принимаю решение заполнить пропуски в столбце rectal_temperature в сцепке с пульсом

horse_data['rectal_temperature'] = horse_data['rectal_temperature'].fillna(horse_data.groupby('pulse')['rectal_temperature'].transform('median'))

#посмотрим, какими значениями заполнились пропуски
horse_data.loc[horse_data['pulse'] == 60].head()
```
